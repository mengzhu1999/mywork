# 三种分布函数

### $$\begin{equation}
f_{\text{FD}}(E) = \frac{1}{e^{\frac{E - \mu}{k_B T}} + 1}
\end{equation}
$$
### $$\begin{equation}
f_{\text{BE}}(E) = \frac{1}{e^{\frac{E - \mu}{k_B T}} - 1}
\end{equation}
$$
### $$\begin{equation}
f_{\text{MB}}(E) = e^{-\frac{E}{k_B T}}
\end{equation}
$$

```python
# Define Fermi-Dirac distribution
def fermi_dirac(E, T, mu):
    # Constants
    k_B = 1.38e-23  # Boltzmann constant (J/K)
    return 1 / (np.exp((E - mu) / (k_B * T)) + 1)

# Define Bose-Einstein distribution
def bose_einstein(E, T, mu):
    # Constants
    k_B = 1.38e-23  # Boltzmann constant (J/K)
    epsilon = 1e-10  # To avoid division by zero
    return 1 / (np.exp((E - mu) / (k_B * T)) - 1 + epsilon)

# Define Maxwell-Boltzmann distribution
def maxwell_boltzmann(E, T):
    # Constants
    k_B = 1.38e-23  # Boltzmann constant (J/K)
    return np.exp(-E / (k_B * T))
```

```python
# Define the plotting function
def plot_distributions(E, T, mu):
    # Calculate distributions
    f_fd = fermi_dirac(E, T, mu)
    f_be = bose_einstein(E, T, mu)
    f_mb = maxwell_boltzmann(E, T)

    # Plot the distributions
    plt.figure(figsize=(8, 6))
    plt.plot(E, f_fd, label='Fermi-Dirac', color='blue')
    plt.plot(E, f_be, label='Bose-Einstein', color='green')
    plt.plot(E, f_mb, label='Maxwell-Boltzmann', color='red')
    plt.xlabel('Energy (J)',ha='center')
    plt.ylabel('Distribution Function',ha='center')
    plt.ylim(0,10)
#     plt.title('Fermi-Dirac, Bose-Einstein, and Maxwell-Boltzmann Distributions')
    plt.title(f'Distributions at T={T},$\mu$={mu}')
    plt.legend()
#     plt.grid(True)
    plt.show()
```

```python
import numpy as np
from scipy.integrate import quad

# 常数
hbar_c = 0.197327  # GeV·fm
pi = np.pi

# 玻色分布函数
def bose_distribution(E, T, mu):
    return 1 / (np.exp((E - mu) / T) - 1)

# 费米分布函数
def fermi_distribution(E, T, mu):
    return 1 / (np.exp((E - mu) / T) + 1)

# 计算强子产额的函数
def hadron_yield(mass, g, T, mu, V, particle_type="boson"):
    # 动量积分：E(p) = sqrt(p^2 + m^2)
    def integrand(p):
        E = np.sqrt(p**2 + mass**2)
        if particle_type == "boson":
            return p**2 * bose_distribution(E, T, mu)
        elif particle_type == "fermion":
            return p**2 * fermi_distribution(E, T, mu)
        else:
            raise ValueError("particle_type should be 'boson' or 'fermion'")
    
    integral, _ = quad(integrand, 0, 100*T)
    return V * g / (2 * pi**2) * integral

# 参数示例
T = 0.155  # 温度，单位GeV
mu_B = 0.03  # 重子化学势，单位GeV
V = 1000  # 系统体积，单位fm^3

# 强子质量和简并度
pion_mass = 0.13957  # π介子的质量，单位GeV
kaon_mass = 0.49367  # K介子的质量，单位GeV
proton_mass = 0.93827  # 质子的质量，单位GeV

g_pion = 3  # π介子的简并度
g_kaon = 4  # K介子的简并度
g_proton = 2  # 质子的简并度

# 计算强子产额
pion_yield = hadron_yield(pion_mass, g_pion, T, 0, V, particle_type="boson")
kaon_yield = hadron_yield(kaon_mass, g_kaon, T, 0, V, particle_type="boson")
proton_yield = hadron_yield(proton_mass, g_proton, T, mu_B, V, particle_type="fermion")

print(f"Pion Yield: {pion_yield:.2f}")
print(f"Kaon Yield: {kaon_yield:.2f}")
print(f"Proton Yield: {proton_yield:.2f}")
```

### 泊松分布
### $$P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

### 泊松分布常用来描述事件在固定时间或空间内的出现次数，特别是当这些事件是独立发生并且在某一段时间或区域内的平均发生次数是已知的，
### 一个常见的例子是电话客服中心接到的呼叫数。如果在某个小时内客服中心平均会接到 10 个电话，我们可以用泊松分布来计算客服中心在某一小时内接到特定数量电话的概率。例如，客服中心接到 7 个电话、10 个电话或 15 个电话的概率，都可以用泊松分布计算

### X 表示事件发生的次数
### k 是我们希望计算的事件发生次数（例如，接到 7 个电话的概率，k=7）
### λ 是已知的平均发生次数。在电话客服中心的例子中，假设每小时接到 10 个电话，则 λ=10
### e 是自然对数的底，约等于 2.718
### 例如，如果平均每小时接到 10 个电话（即 λ=10），我们可以计算接到正好 7 个电话的概率：
## 泊松分布例子：
### $$P(X = 7) = \frac{10^7 e^{-10}}{7!}$$

```python
from scipy.stats import poisson

def plot_poisson_distribution(lambda_values):
    """
    绘制不同 λ 值的泊松分布图像。

    参数:
    lambda_values (list): λ 值的列表，每个 λ 值代表平均发生次数。
    """
    k_values = np.arange(0, 20)  # 事件次数的范围

    # 创建图像
    plt.figure(figsize=(10, 6))
    for lam in lambda_values:
        # 计算泊松分布的概率质量函数
        pmf_values = poisson.pmf(k_values, lam)
        plt.plot(k_values, pmf_values, marker='o', linestyle='-', label=f'λ = {lam}')

    # 添加标签和图例
    plt.title("Poisson Distribution")
    plt.xlabel("Number of Events (k)", ha='center')
    plt.ylabel("Probability", ha='center')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
# 示例调用
lambda_values = [1, 4, 10]
plot_poisson_distribution(lambda_values)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 初始条件
rho_L, u_L, p_L = 1.0, 0.0, 1.0  # 左侧密度、速度、压力
rho_R, u_R, p_R = 0.125, 0.0, 0.1  # 右侧密度、速度、压力
gamma = 1.4  # 比热比

# 定义稀疏波速度函数
def sound_speed(p, rho, gamma):
    return np.sqrt(gamma * p / rho)

# 求解稀疏波和激波的速度
c_L = sound_speed(p_L, rho_L, gamma)
c_R = sound_speed(p_R, rho_R, gamma)

# 空间网格和时间
x = np.linspace(-1, 1, 500)
t = 0.25  # 选择一个时间点

# 定义不同区域的解
rho = np.where(x < -c_L * t, rho_L, np.where(x > c_R * t, rho_R, (rho_L + rho_R) / 2))
u = np.where(x < -c_L * t, u_L, np.where(x > c_R * t, u_R, (u_L + u_R) / 2))
p = np.where(x < -c_L * t, p_L, np.where(x > c_R * t, p_R, (p_L + p_R) / 2))

# 绘制结果
plt.figure(figsize=(10, 6))
plt.subplot(3, 1, 1)
plt.plot(x, rho, label='Density')
plt.ylabel('Density')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(x, u, label='Velocity')
plt.ylabel('Velocity')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(x, p, label='Pressure')
plt.ylabel('Pressure')
plt.legend()

plt.xlabel('Position')
plt.suptitle(f'Riemann Solution at t = {t}')
plt.show()
```

### $$H(\eta_s) = \exp \left[ -\frac{(|\eta_s| - \eta_w)^2}{2\sigma_\eta^2} \theta(|\eta_s| - \eta_w) \right]$$


```python
# 定义更新后的 H(η_s) 函数
def H(eta_s, eta_w, sigma_eta):
    # 使用阶跃函数 θ(|η_s| - η_w)
    theta = np.heaviside(np.abs(eta_s) - eta_w, 0)
    # 指数函数，其中包含绝对值符号
    return np.exp(-((np.abs(eta_s) - eta_w) ** 2) / (2 * sigma_eta ** 2) * theta)

# 绘图函数
def plot_H(eta_w=1.0, sigma_eta=0.5, eta_range=(-3, 3), num_points=1000):
    eta_s = np.linspace(eta_range[0], eta_range[1], num_points)
    H_values = H(eta_s, eta_w, sigma_eta)
    
    plt.figure(figsize=(8, 6))
    plt.plot(eta_s, H_values, label=fr"$H(\eta_s)$ with $\eta_w={eta_w}$, $\sigma_\eta={sigma_eta}$")
    plt.xlabel(r"$\eta_s$", ha='center')
    plt.ylabel(r"$H(\eta_s)$", ha='center')
    plt.title(r"$H(\eta_s) = \exp\left[-\frac{(|\eta_s| - \eta_w)^2}{2\sigma_\eta^2} \theta(|\eta_s| - \eta_w)\right]$")
    plt.legend()
    plt.grid(True)
    plt.show()

# 示例用法
plot_H(eta_w=1.0, sigma_eta=1)
```

### 负二项分布


```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb  # 用于计算组合数

def negative_binomial_pmf(n, k, avg_n):
    """
    计算负二项分布 PMF (基于公式 P_n(average_n, k))

    参数:
    n: 观测到的事件数（失败次数）
    k: 成功次数
    avg_n: 平均事件数（\overline{n}）

    返回:
    PMF 值
    """
    # 计算组合数
    coeff = comb(n + k - 1, k - 1)
    # 计算概率项
    prob_failure = (avg_n / (1 + avg_n))**n
    prob_success = 1 / (1 + avg_n)**k
    # 返回 PMF
    return coeff * prob_failure * prob_success


def plot_negative_binomial(avg_n, k, max_n=20):
    """
    绘制负二项分布 PMF 图像

    参数:
    avg_n: 平均事件数 (\overline{n})
    k: 成功次数
    max_n: 最大绘制的 n 值
    """
    # 生成 n 的取值
    n_values = np.arange(0, max_n + 1)
    # 计算对应的 PMF 值
    pmf_values = [negative_binomial_pmf(n, k, avg_n) for n in n_values]
    
    # 绘图
    plt.figure(figsize=(8, 6))
    plt.stem(n_values, pmf_values, basefmt=" ", use_line_collection=True)
    plt.title(f"Negative Binomial Distribution (avg_n={avg_n}, k={k})", fontsize=14)
    plt.xlabel("n (observed failures)", fontsize=12)
    plt.ylabel("PMF", fontsize=12)
    plt.grid(True, linestyle="--", alpha=0.7)
    plt.show()
```

```python
# 示例 1: 改变平均事件数 avg_n
plot_negative_binomial(avg_n=2, k=5, max_n=20)  # avg_n=2, k=5
plot_negative_binomial(avg_n=5, k=5, max_n=20)  # avg_n=5, k=5
plot_negative_binomial(avg_n=10, k=5, max_n=20)  # avg_n=10, k=5

# 示例 2: 改变成功次数 k
plot_negative_binomial(avg_n=5, k=2, max_n=20)  # avg_n=5, k=2
plot_negative_binomial(avg_n=5, k=10, max_n=20)  # avg_n=5, k=10
```

```python
def plot_spherical_harmonic_2d(l, m):
    """
    绘制给定量子数 l 和 m 的球谐函数 Y_l^m 的二维角度分布图（θ-φ 平面）。
    
    参数:
    l (int): 主量子数 (非负整数)
    m (int): 磁量子数 (-l <= m <= l)
    
    返回:
    None
    """
    # 创建角度网格
    theta = np.linspace(0, np.pi, 200)  # 极角 θ: 0 到 π
    phi = np.linspace(0, 2 * np.pi, 200)  # 方位角 φ: 0 到 2π
    theta, phi = np.meshgrid(theta, phi)  # 创建二维网格
    
    # 计算球谐函数
    Y_lm = sph_harm(m, l, phi, theta)
    Y_lm_magnitude = np.abs(Y_lm)  # 提取幅值（模）
    
    # 绘制二维分布图
    plt.figure(figsize=(8, 6))
    plt.pcolormesh(phi, theta, Y_lm_magnitude, cmap="viridis", shading="auto")
    plt.colorbar(label="|Yₗᵐ(θ, φ)|")  # 添加颜色条，表示幅值
    
    # 设置轴标签和标题
    plt.xlabel("φ (Azimuthal Angle)", fontsize=12)
    plt.ylabel("θ (Polar Angle)", fontsize=12)
    plt.title(f"2D Angular Distribution: $Y_{{{l},{m}}}$", fontsize=15)
    
    # 显示图形
#     plt.tight_layout()
#     plt.show()
```

```python
def plot_spherical_harmonic_no_axes_subplots(params,save_fig=False):
    """
    在一个画布上绘制多个球谐函数 Y_l^m 的 3D 图像（无坐标轴、网格）。
    
    参数:
    params (list of tuples): 每个元组包含 (l, m)，表示需要绘制的量子数组合。
                             例如 [(2, 2), (3, 2), (3, 3)]。
    
    返回:
    None
    """
    # 设置画布大小和子图网格
    num_plots = len(params)  # 确定子图数量
    cols = 4  # 每行显示的图像数量
    rows = (num_plots + cols - 1) // cols  # 根据图像数量动态调整行数

    fig = plt.figure(figsize=(12, 6))  # 调整画布大小
    
    for idx, (l, m) in enumerate(params, start=1):
        # 创建子图
        ax = fig.add_subplot(rows, cols, idx, projection='3d')

        # 创建球坐标范围
        theta = np.linspace(0, np.pi, 100)  # 极角 θ: 0 到 π
        phi = np.linspace(0, 2 * np.pi, 100)  # 方位角 φ: 0 到 2π
        theta, phi = np.meshgrid(theta, phi)  # 创建二维网格
        
        # 计算球谐函数
        Y_lm = sph_harm(m, l, phi, theta).real  
        Y_lm_magnitude = np.abs(Y_lm)**2  # 提取幅值方（模方）
        
        # 映射到笛卡尔坐标
        x = Y_lm_magnitude * np.sin(theta) * np.cos(phi)
        y = Y_lm_magnitude * np.sin(theta) * np.sin(phi)
        z = Y_lm_magnitude * np.cos(theta)

        # 绘制球谐函数的 3D 表面
        ax.plot_surface(
            x, y, z, 
            rstride=1, cstride=1, 
            facecolors=plt.cm.viridis(Y_lm_magnitude / Y_lm_magnitude.max()), 
            edgecolor='none'
        )
        
        # 移除坐标轴和背景
        ax.set_axis_off()

        # 设置视角为沿 y 轴的方向
#         ax.view_init(elev=0, azim=90)

        # 添加标题
        ax.set_title(f"$|Y_{{{l},{m}}}|^2$", fontsize=18, y=0.95)
        if save_fig:
            plt.savefig("./d_Au_work/fig/spherical_harmonic.pdf",bbox_inches="tight")
            
    # 添加全局颜色条
#     cbar_ax = fig.add_axes([0.92, 0.15, 0.02, 0.7])  # 定义颜色条的位置 [left, bottom, width, height]
#     norm = plt.cm.ScalarMappable(cmap="viridis")
#     cbar = fig.colorbar(norm, cax=cbar_ax)
#     cbar.set_label("|$Y_l^m(\\theta, \\phi)|^2$ (Magnitude)", fontsize=12)
    # 调整子图间距
#     plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import sph_harm
from matplotlib.colors import Normalize, TwoSlopeNorm, LinearSegmentedColormap
import matplotlib.cm as cm

def plot_spherical_harmonic_no_axes_subplots(params, save_fig=False):
    num_plots = len(params)
    cols = 5
    rows = (num_plots + cols - 1) // cols
    fig = plt.figure(figsize=(15, 3 * rows))

    for idx, param in enumerate(params, start=1):
        ax = fig.add_subplot(rows, cols, idx, projection='3d')
        theta = np.linspace(0, np.pi, 100)
        phi = np.linspace(0, 2 * np.pi, 100)
        theta, phi = np.meshgrid(theta, phi)

        if isinstance(param[0], tuple):  # 输入是 ((l1,m1), (l2,m2))
            (l1, m1), (l2, m2) = param
            Y1 = sph_harm(m1, l1, phi, theta)
            Y2 = sph_harm(m2, l2, phi, theta)
            Y_lm = Y1 * Y2
            r = Y_lm.real  # 取实部
            title = f"$Y_{{{l1},{m1}}} Y_{{{l2},{m2}}}$"

            # 提取 viridis 的蓝色
            viridis_norm = Normalize(vmin=0, vmax=np.max(r))
            viridis_cmap = cm.viridis
            blue_color = viridis_cmap(viridis_norm(0.1))  # 提取 norm=0.1 对应的颜色

            # 定义紫色
            purple_color = (0.5, 0, 0.5)  # 深紫色

            # 创建自定义颜色映射
            custom_cmap = LinearSegmentedColormap.from_list('custom_cmap', [purple_color, blue_color])

            # 使用 TwoSlopeNorm 归一化
            norm = TwoSlopeNorm(vmin=np.min(r), vcenter=0, vmax=np.max(r))
            colors = custom_cmap(norm(r))  # 使用自定义颜色映射

        else:  # 输入是 (l, m)
            l, m = param
            Y_lm = sph_harm(m, l, phi, theta)
            r = np.abs(Y_lm)**2  # 取模的平方
            title = f"$|Y_{{{l},{m}}}|^2$"

            # 使用 Normalize 归一化
            norm = Normalize(vmin=0, vmax=np.max(r))
            colors = cm.viridis(norm(r))  # 使用 viridis 颜色映射

        # 转换为笛卡尔坐标
        x = r * np.sin(theta) * np.cos(phi)
        y = r * np.sin(theta) * np.sin(phi)
        z = r * np.cos(theta)

        ax.plot_surface(
            x, y, z,
            facecolors=colors,
            rstride=1, cstride=1,
            linewidth=0,
            antialiased=False,
            shade=False
        )

        max_range = np.max(np.abs(r))
        ax.set_xlim([-max_range, max_range])
        ax.set_ylim([-max_range, max_range])
        ax.set_zlim([-max_range, max_range])

        ax.set_box_aspect([1,1,1])
        ax.set_title(title, fontsize=20)
        ax.set_axis_off()

    if save_fig:
        plt.savefig("../d_Au_work/fig/spherical_harmonics_one_uni.pdf", bbox_inches="tight")
    plt.tight_layout()
    plt.show()

# 测试
params = [(0, 0), (2, 2), (2, 1), (2, 0), ((0, 0), (2, 0))]
plot_spherical_harmonic_no_axes_subplots(params,save_fig=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import sph_harm
from matplotlib.colors import TwoSlopeNorm

def plot_spherical_harmonic_no_axes_subplots(params, save_fig=False):
    """
    绘制球谐函数（或两个球谐函数的乘积）的 3D 图像。
    
    参数:
    params (list): 每个元素可以是：
                   - 元组 (l, m): 绘制单个球谐函数 |Y_l^m|^2
                   - 元组 ((l1, m1), (l2, m2)): 绘制 |Y_{l1,m1} * Y_{l2,m2}|^2
                   例如：
                   [(2, 2), (3, 1), ((0, 0), (2, 0))]  # 最后一个表示 Y_{00} * Y_{20}
    save_fig (bool): 是否保存图像。
    """
    num_plots = len(params)
    cols = 5  # 每行 5 列
    rows = (num_plots + cols - 1) // cols
    fig = plt.figure(figsize=(15, 3 * rows))  # 动态调整画布高度

    for idx, param in enumerate(params, start=1):
        ax = fig.add_subplot(rows, cols, idx, projection='3d')
        theta = np.linspace(0, np.pi, 100)
        phi = np.linspace(0, 2 * np.pi, 100)
        theta, phi = np.meshgrid(theta, phi)

        # 判断输入是单个球谐函数还是乘积
        if isinstance(param[0], tuple):  # 输入是 ((l1,m1), (l2,m2))
            (l1, m1), (l2, m2) = param
            Y1 = sph_harm(m1, l1, phi, theta)
            Y2 = sph_harm(m2, l2, phi, theta)
            Y_lm = Y1 * Y2
            r = Y_lm.real  # 取实部
            title = f"$Y_{{{l1},{m1}}} \\cdot Y_{{{l2},{m2}}}$"  # 使用双反斜杠
        else:  # 输入是 (l, m)
            l, m = param
            Y_lm = sph_harm(m, l, phi, theta)
            r = np.abs(Y_lm)**2  # 取模的平方
            title = f"$|Y_{{{l},{m}}}|^2$"

        # 转换为笛卡尔坐标
        x = r * np.sin(theta) * np.cos(phi)
        y = r * np.sin(theta) * np.sin(phi)
        z = r * np.cos(theta)

        # 使用 TwoSlopeNorm 归一化
        norm = TwoSlopeNorm(vmin=-0.1, vcenter=0.1, vmax=0.4)
        colors = plt.cm.RdBu(norm(r))  # 使用红蓝颜色映射来区分正负值

        ax.plot_surface(
            x, y, z,
            facecolors=colors,
            rstride=1, cstride=1,
            linewidth=0,
            antialiased=False,
            shade=False
        )

        # 手动设置坐标轴范围
        max_range = np.max(np.abs(r))  # 根据实部的最大绝对值设置范围
        ax.set_xlim([-max_range, max_range])
        ax.set_ylim([-max_range, max_range])
        ax.set_zlim([-max_range, max_range])

        ax.set_box_aspect([1,1,1])  # 保持盒子比例
        ax.set_title(title, fontsize=16)
        ax.set_axis_off()  # 不显示坐标轴

    if save_fig:
        plt.savefig("../d_Au_work/fig/spherical_harmonics_all_uni.pdf", bbox_inches="tight")
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import sph_harm
from matplotlib.colors import Normalize

def plot_spherical_harmonic_dual_rows(params, save_fig=False):
    """
    绘制球谐函数的概率幅和概率密度分布，分上下两行显示。
    
    参数:
    params (list of tuples): 每个元组 (l, m) 表示量子数组合，例如 [(2,0), (2,1), (3,1), (3,2)]。
    save_fig (bool): 是否保存图像。
    """
    num_plots = len(params)
    cols = min(10, num_plots)  # 每行最多4个子图
    rows = 2  # 固定两行：上行概率幅，下行概率密度

    fig = plt.figure(figsize=(15, 8))  # 调整画布大小

    # 全局颜色范围（基于所有数据）
    global_amp_max = 0
    global_density_max = 0
    all_amp_data = []
    all_density_data = []

    # 预计算所有数据
    for l, m in params:
        theta = np.linspace(0, np.pi, 100)
        phi = np.linspace(0, 2 * np.pi, 100)
        theta_grid, phi_grid = np.meshgrid(theta, phi)
        
        Y_lm = sph_harm(m, l, phi_grid, theta_grid) # 取实部
        prob_amp = np.abs(Y_lm)       # 概率幅 |Y_lm|
        prob_density = prob_amp ** 2  # 概率密度 |Y_lm|^2
        
        all_amp_data.append(prob_amp)
        all_density_data.append(prob_density)
        global_amp_max = max(global_amp_max, prob_amp.max())
        global_density_max = max(global_density_max, prob_density.max())

    # 统一颜色映射
    norm_amp = Normalize(vmin=0, vmax=global_amp_max)
    norm_density = Normalize(vmin=0, vmax=global_density_max)

    # 绘制概率幅（上行）
    for idx, ((l, m), prob_amp) in enumerate(zip(params, all_amp_data), start=1):
        ax = fig.add_subplot(rows, cols, idx, projection='3d')
        
        # 笛卡尔坐标
        x = prob_amp * np.sin(theta_grid) * np.cos(phi_grid)
        y = prob_amp * np.sin(theta_grid) * np.sin(phi_grid)
        z = prob_amp * np.cos(theta_grid)
        
        # 绘图
        ax.plot_surface(
            x, y, z,
            facecolors=plt.cm.viridis(norm_amp(prob_amp)),
            edgecolor='none',
            rstride=1, cstride=1
        )
        ax.set_title(f"$|Y_{{{l},{m}}}|$", fontsize=20, y=0.95)
        ax.set_axis_off()

    # 绘制概率密度（下行）
    for idx, ((l, m), prob_density) in enumerate(zip(params, all_density_data), start=cols+1):
        ax = fig.add_subplot(rows, cols, idx, projection='3d')
        
        x = prob_density * np.sin(theta_grid) * np.cos(phi_grid)
        y = prob_density * np.sin(theta_grid) * np.sin(phi_grid)
        z = prob_density * np.cos(theta_grid)
        
        ax.plot_surface(
            x, y, z,
            facecolors=plt.cm.viridis(norm_density(prob_density)),
            edgecolor='none',
            rstride=1, cstride=1
        )
        ax.set_title(f"$|Y_{{{l},{m}}}|^2$", fontsize=20, y=0.95)
        ax.set_axis_off()

#     # 添加全局颜色条
#     cbar_ax_amp = fig.add_axes([0.15, 0.45, 0.7, 0.02])  # 上行颜色条位置
#     cbar_amp = fig.colorbar(
#         ScalarMappable(norm=norm_amp, cmap='viridis'),
#         cax=cbar_ax_amp,
#         orientation='horizontal',
#         label='Probability Amplitude ($|Y_{l,m}|$)'
#     )

#     cbar_ax_density = fig.add_axes([0.15, 0.25, 0.7, 0.02])  # 下行颜色条位置
#     cbar_density = fig.colorbar(
#         ScalarMappable(norm=norm_density, cmap='viridis'),
#         cax=cbar_ax_density,
#         orientation='horizontal',
#         label='Probability Density ($|Y_{l,m}|^2$)'
#     )

    plt.tight_layout(rect=[0, 0.3, 1, 1])  # 调整布局避免重叠
    if save_fig:
        plt.savefig("spherical_harmonics_dual_rows.pdf", bbox_inches="tight")
    plt.show()

# 示例调用
params = [(2,2), (2,1), (2,0), (2,-1), (2,-2)]  # 可扩展至最多4个参数
# params = [(0,0)]  # 可扩展至最多4个参数
# params = [(1,1), (1,0), (1,-1)]  # 可扩展至最多4个参数
# params = [(3,3), (3,2), (3,1), (3,0), (3,-1), (3,-2), (3,-3)]  # 可扩展至最多4个参数


plot_spherical_harmonic_dual_rows(params, save_fig=False)
```

```python
def plot_multiple_spherical_harmonics_2d(params):
    """
    绘制多个球谐函数 Y_l^m 的二维角度分布图在同一个画布中。

    参数:
    params (list of tuples): 每个元组包含 (l, m)，表示需要绘制的量子数组合。
                             例如 [(2, 2), (2, 1), (2, 0), (1, 0)]。

    返回:
    None
    """
    cols = 4  # 每行显示的子图数量
    rows = 1  # 子图的行数
    
    fig, axes = plt.subplots(rows, cols, figsize=(14, 3))  # 创建子图布局
    
    for idx, (l, m) in enumerate(params):
        ax = axes[idx]  # 获取当前子图的Axes对象
        
        # 创建角度网格
        theta = np.linspace(0, np.pi, 200)  # 极角 θ: 0 到 π
        phi = np.linspace(0, 2 * np.pi, 200)  # 方位角 φ: 0 到 2π
        theta, phi = np.meshgrid(theta, phi)  # 创建二维网格

        # 计算球谐函数
        Y_lm = sph_harm(m, l, phi, theta)
        Y_lm_magnitude = np.abs(Y_lm)  # 提取幅值（模）
        
        if l == 0 and m == 0:  # 对 Y_{0,0} 特殊处理
            Y_lm_magnitude = np.ones_like(theta)  # 全部设为最大值（均匀分布）
            im = ax.pcolormesh(
                phi, theta, Y_lm_magnitude, 
                cmap="viridis", shading="auto", 
                vmin=0, vmax=1  # 确保使用 viridis 中的最大颜色
            )
        else:
            # 正常计算其他球谐函数
            Y_lm = sph_harm(m, l, phi, theta)
            Y_lm_magnitude = np.abs(Y_lm)**2
            im = ax.pcolormesh(
                phi, theta, Y_lm_magnitude, 
                cmap="viridis", shading="auto"
            )
        ax.set_xlabel("$\phi$", fontsize=20)
        ax.set_ylabel(r"$\theta$", fontsize=20)
    
    # 删除多余的空子图（如果子图数量少于布局大小）
    for idx in range(len(params), rows * cols):
        fig.delaxes(axes[idx])
    
    # 添加全局颜色条
#     cbar_ax = fig.add_axes([0.92, 0.15, 0.02, 0.7])  # [left, bottom, width, height]
#     norm = plt.cm.ScalarMappable(cmap="viridis")
#     fig.colorbar(norm, cax=cbar_ax, label="$|Y_{l,m}(θ, φ)|$")

    # 调整布局和间距
    plt.subplots_adjust(left=0.0, right=0.9, top=0.95, bottom=0.05, wspace=0.3, hspace=0.3)
    plt.show()

```

```python
from scipy.stats import gamma
import numpy as np
import matplotlib.pyplot as plt

# 定义形状参数 k
k = 1  # 你可以根据需要改变这个值

# 生成伽马分布的随机样本
samples = gamma.rvs(a=k, scale=1/k, size=1000)

# 绘制直方图和概率密度函数
count, bins, ignored = plt.hist(samples, 30, density=True, alpha=0.5, color='g')

# 绘制理论概率密度函数
x = np.linspace(0, max(samples), 100000)
pdf = gamma.pdf(x, k, scale=1/k)
plt.plot(x, pdf, 'r-', lw=2)

plt.title('Gamma Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma
import mplhep as hep
hep.style.use('ATLAS')

# 设置不同的k值
k_values = [0.5, 1, 2, 4, 6]

# 定义x的范围，缩小到0到5
x = np.linspace(0, 5, 1000)

# 绘制每个k值对应的伽马分布PDF
plt.figure(figsize=(12, 8))  # 增加图像尺寸
for k in k_values:
    # 计算PDF
    pdf = gamma.pdf(x, a=k, scale=1)
    # 绘制PDF，使用不同的线条样式和透明度
    plt.plot(x, pdf, label=f'k={k}', linestyle='-', linewidth=2, alpha=0.7)

# 添加图例
plt.legend()

# 添加标题和轴标签
plt.title('Gamma Distribution for Different k Values')
plt.xlabel('Value')
plt.ylabel('Probability Density')

# 设置x轴和y轴的限制
plt.xlim(0, 5)
plt.ylim(0, max(gamma.pdf(0.1, a=k, scale=1) for k in k_values) * 1.1)

# 显示图形
plt.show()
```

```python
def plot_gamma_distribution(k_values):
    hep.style.use('ATLAS')

    # 定义x的范围，缩小到0到5
    x = np.linspace(0, 30, 1000)

    # 绘制每个k值对应的伽马分布PDF
    plt.figure(figsize=(12, 8))  # 增加图像尺寸
    for k in k_values:
        # 计算PDF
        pdf = gamma.pdf(x, a=k, scale=1)
        # 绘制PDF，使用不同的线条样式和透明度
        plt.plot(x, pdf, label=f'k={k}', linestyle='-', linewidth=2, alpha=0.7)

    # 添加图例
    plt.legend()

    # 添加标题和轴标签
    plt.title('Gamma Distribution for Different k Values')
    plt.xlabel('Value')
    plt.ylabel('Probability Density')

    # 设置x轴和y轴的限制
#     plt.xlim(0, 10)
    plt.ylim(0, max(gamma.pdf(0.1, a=k, scale=1) for k in k_values) * 1.1)

    # 显示图形
    plt.show()

```

```python
# 使用示例
k_values = [0.5, 1, 4, 6, 20]
plot_gamma_distribution(k_values)
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def calculate_pseudorapidity(theta):
    """
    计算赝快度（pseudorapidity）。

    参数:
    theta -- 粒子的极角（以弧度为单位）

    返回:
    赝快度的值
    """
    # 避免除以零和对数中的负无穷
    if np.isclose(theta, 0):
        return np.inf
    elif np.isclose(theta, np.pi):
        return -np.inf
    else:
        return -np.log(np.tan(theta / 2))

# 生成极角的值，从0到π，避免0和π
theta_values = np.linspace(0.0001, np.pi - 0.0001, 1000)  # 1000个点，从0.0001到π-0.0001
eta_values = [calculate_pseudorapidity(theta) for theta in theta_values]

# 绘制赝快度与极角的关系图
plt.figure(figsize=(10, 6))
plt.plot(theta_values, eta_values, label=r'Pseudorapidity $\eta$ vs Theta $\theta$')
plt.xlabel(r'$\theta$ (radians)')  # 使用LaTeX格式化X轴标签
plt.ylabel(r'$\eta$')  # 使用LaTeX格式化Y轴标签
plt.title(r'Pseudorapidity $\eta$ vs Theta $\theta$')  # 使用LaTeX格式化标题
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 设置x和y的范围
x = np.linspace(-5, 5, 100)  # x轴范围
y = np.linspace(-5, 5, 100)  # y轴范围

# 生成网格点
X, Y = np.meshgrid(x, y)

# 定义能量密度函数（二维高斯分布）
sigma = 1.0  # 高斯分布的标准差
energy_density = np.exp(-(X**2 + Y**2) / (2 * sigma**2))

# 创建图形和3D坐标轴
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# 绘制三维曲面图
surf = ax.plot_surface(X, Y, energy_density, cmap='viridis', edgecolor='none')

# 添加颜色条
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

# 设置坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Energy Density')

# 设置标题
ax.set_title('3D Energy Density Distribution')

# 显示图形
plt.show()
```
