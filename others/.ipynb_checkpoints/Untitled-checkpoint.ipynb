# 并行加速

```python
from concurrent.futures import ThreadPoolExecutor, as_completed

def square_number(n):
    return n * n

numbers = [1, 2, 3, 4, 5]
with ThreadPoolExecutor(max_workers=5) as executor:
    futures = [executor.submit(square_number, num) for num in numbers]
    results = [f.result() for f in as_completed(futures)]
print(results)
```

```python
def plateau_function(eta=np.linspace(-6.9, 6.9, 121)):
    """Compute the plateau function based on pseudorapidity."""
    eta_gw=0.5
    eta_flat=1
    heta = np.ones_like(eta)
    fall_off = np.abs(eta) > eta_flat
    eta_fall = np.abs(eta[fall_off])
    heta[fall_off] = np.exp(-(eta_fall - eta_flat)**2 / (2.0 * eta_gw**2))
    return heta
```

```python
x = eta=np.linspace(-6.9, 6.9, 121)
plt.plot(x,plateau_function(eta=np.linspace(-6.9, 6.9, 121)) )
```

```python
def gaussian_function(x=0, center=0, width=1):
    """Compute a Gaussian function."""
    return np.exp(-((x - center)**2) / (2 * width**2))
```

```python
x = eta=np.linspace(-6.9, 6.9, 121)
plt.plot(x,gaussian_function(x=x, center=-2, width=1) )
```

```python
A = table[:,1]
a = table[:,2]
B = table[:,3]
b = table[:,4]
A
```

```python
def u(r):
    avu = 0
    for i in range(len(A)):
        avu += r**1.5 * A[i] * np.exp(- a[i] * r**3)
    return avu
```

```python
def w(r):
    avw = 0
    for i in range(len(B)):
        avw += r * B[i] * np.exp(- b[i] * r**3)
    return avw
```

```python
def P_S(r):
    avu = 0
    for i in range(len(A)):
        avu += r**1.5 * A[i] * np.exp(- a[i] * r**3)
    return avu**2
```

```python
def P_D(r):
    avw = 0
    for i in range(len(B)):
        avw += r * B[i] * np.exp(- b[i] * r**3)
    return avw**2
```

### 定义三个部分的概率密度分布

```python
def p_u(r,theta):
    
    return u(r)**2/(4*np.pi) * np.sin(theta)
```

```python
def p_uw(r,theta):
    
    return u(r) * w(r)/(4*np.pi) * ( np.sqrt(1/2) * (3 * np.cos(theta)**2 - 1)) * np.sin(theta)
```

```python
def p_w(r,theta):
    
    return w(r)**2/(4*np.pi) * (9/8 * np.sin(theta)**4 
                                 + 9/4 * np.sin(theta)**2 * np.cos(theta)**2
                                 + 1/8 * (3 * np.cos(theta)**2 - 1)**2) * np.sin(theta)
```

```python
#定义总的概率密度函数
def p(r,theta):
    return p_u(r,theta) + p_w(r,theta)
```

```python
def rejection_sampling(pdf, n=1000, ymax=1, xmin=0, 
                      xmax_theta=np.pi, xmax_phi=2*np.pi, xmax_r=25,
                      batch_size=10000):
    '''向量化采样（单进程，但利用NumPy加速）'''
    samples_r, samples_theta, samples_phi = [], [], []
    pbar = tqdm(total=n, desc="采样进度")
    
    while len(samples_r) < n:
        # 批量生成候选点
        r = xmin + (xmax_r - xmin) * np.random.rand(batch_size)
        theta = xmin + (xmax_theta - xmin) * np.random.rand(batch_size)
        phi = xmin + (xmax_phi - xmin) * np.random.rand(batch_size)
        y = ymax * np.random.rand(batch_size)
        
        # 批量接受
        mask = y < pdf(r, theta)
        samples_r.extend(r[mask])
        samples_theta.extend(theta[mask])
        samples_phi.extend(phi[mask])
        pbar.update(len(r[mask]))
    
    pbar.close()
    return (
        np.array(samples_r[:n], dtype=np.float32),
        np.array(samples_theta[:n], dtype=np.float32),
        np.array(samples_phi[:n], dtype=np.float32)
    )
```

```python
DWF_d_r_theta_phi_1000000 = rejection_sampling(pdf=p, n=10000000, ymax=1, xmin=0, 
                      xmax_theta=np.pi, xmax_phi=2*np.pi, xmax_r=25,
                      batch_size=10000)
```
