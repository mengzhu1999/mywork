## 读取实验 pT spectra 数据

```python
def read_hepdata_with_pandas(file_path):
    data = {}
    current_centrality = None
    
    with open(file_path, 'r') as file:
        for line in file:
            line = line.strip()
            if not line:
                continue
            if line.startswith('#:'):
                if 'centrality' in line:
                    parts = [p.strip() for p in line.split(',') if p.strip()]
                    if len(parts) >= 2:
                        current_centrality = parts[1]
                        data[current_centrality] = []
            elif current_centrality is not None:
                parts = line.split(',')
                if len(parts) >= 6:
                    try:
                        pT = float(parts[0])
                        pT_low = float(parts[1])
                        pT_high = float(parts[2])
                        yield_val = float(parts[3])
                        stat_err_plus = float(parts[4])
                        stat_err_minus = float(parts[5])
                        
                        data[current_centrality].append({
                            'pT': pT,
                            'pT_low': pT_low,
                            'pT_high': pT_high,
                            'yield': yield_val,
                            'stat_err_plus': stat_err_plus,
                            'stat_err_minus': stat_err_minus
                        })
                    except ValueError:
                        continue
    
    # 将数据转换为DataFrame
    dfs = {}
    for centrality, entries in data.items():
        dfs[centrality] = pd.DataFrame(entries)
    return dfs

```

## 计算模拟数据 pT spectra

```python
def get_pt_spectrum_data(m=0, n=20, pid=321, nsampls=4000, pt_bins=30, pt_range=(0, 3)):
    """
    获取 pT 谱数据（使⽤ np.histogram 的 weights 直接计算均值）
    
    参数:
    - m, n: 路径参数
    - pid: 粒⼦ PDG 编码（默认 321，K⁺）
    - nsampls: 抽样次数（默认 2000）
    - pt_bins: pT 分 bin 数量（默认 30）
    - pt_range: pT 范围（默认 0-5 GeV）
    
    返回:
    - dict: 包含 bin_centers, bin_means, bin_stds, bin_counts, bin_edges, dpt
    """
#     base_path = f"/media/mengzhu/west16/d_Au_work/d_Au_DWF/hydro_result_n_BC/ebe_mc_nsampling_{nsampls}/plat_1p3_gw_1p3/entropy3D_y0_4p0_sigma_2p5_L_factor_2p0_R_factor_2p0_P_factor_5p5_alpha_1p0_beta_0p35/dAu200"
    base_path = "/media/mengzhu/west16/Au_Au_work/hydro_result_n_BC_pT_spectra/Tfrz_128Mev_ETAOS_0p16_tau0p6/ebe_mc_nsampling_2000/plat_1p3_gw_1p3/"\
                "entropy3D_y0_4p0_sigma_2p5_L_factor_2p0_R_factor_2p0_P_factor_5p5_alpha_1p0_beta_0p47/AuAu200/"  
    event_start, event_end = 0, 1000
    all_pt = []
    all_dN = []
    # 1. 加载数据
    for event in range(event_start, event_end ):
        file_path = os.path.join(base_path, f"{m}_{n}/event{event}/dN_over_2pidYptdpt_mc_{pid}.dat")
        if os.path.exists(file_path):
            try:
                data = np.loadtxt(file_path)
                all_pt.append(data[:, 0]) # pT
                all_dN.append(data[:, 1]) # dN/dpT
            except Exception as e:
                print(f"Error loading event {event}: {e}")
    # 2. 合并数据
    all_pt_flat = np.concatenate(all_pt)
    all_dN_flat = np.concatenate(all_dN)
    
    # 3. 分 bin 统计
    dN, pt_edges = np.histogram(all_pt_flat, bins=pt_bins, range=pt_range)
    dpt = pt_edges[1:] - pt_edges[:-1]                # 计算 bin 宽度
    pt_centers = 0.5 * (pt_edges[:-1] + pt_edges[1:]) # 计算 bin 中心
    
    # 4. 关键修改：⽤ weights 直接计算均值
    sum_dN, _ = np.histogram(all_pt_flat, bins=pt_edges, weights=all_dN_flat)
    counts, _ = np.histogram(all_pt_flat, bins=pt_edges)
    
    # 5. 仅仅计算 bins 中数据点的部分，因为 high pT 时，可能 bins 中并不存在粒⼦
    valid_bins = counts > 0
    bin_means = np.zeros_like(counts, dtype=float)
    bin_stds = np.zeros_like(counts, dtype=float)
    
    bin_means[valid_bins] = sum_dN[valid_bins] / counts[valid_bins]
    bin_stds[valid_bins] = bin_means[valid_bins] / np.sqrt(counts[valid_bins])
    bin_means = np.where(counts > 0, sum_dN / counts, 0)             # 均值 = sum(dN) / counts
    bin_stds = np.where(counts > 0, bin_means / np.sqrt(counts), 0)  # 误差 = mean / sqrt(N)
    return {
        'bin_centers': pt_centers,
        'bin_means': bin_means,
        'bin_stds': bin_stds,
        'bin_counts': counts,
        'bin_edges': pt_edges,
        'dpt': dpt,
        }
```

## 实验与模拟比较

```python
from matplotlib.lines import Line2D

def plot_particle_multi_centrality(pid=211, centralities=["0_5", "5_10", "10_15"], 
                                  pt_range=(0, 3), ylog=True, scale_factors=None):
    """
    绘制指定粒子在不同中心度下的实验与模拟数据对比
    
    参数:
        pid : 粒子ID (211:π⁺, 321:K⁺, 2212:p)
        centralities : 要对比的中心度列表 (如 ["0_5", "5_10"])
        pt_range : pT范围 (GeV)
        ylog : 是否使用对数y轴
        scale_factors : 各中心度的缩放因子 (如 [1.0, 0.5, 0.2])
    """
    # 实验数据映射
    exp_data_mapping = {
        211: {  # π⁺
            "0_5": pion_plus_0_5, "5_10": pion_plus_5_10, "10_15": pion_plus_10_15,
            "15_20": pion_plus_15_20, "20_30": pion_plus_20_30, "30_40": pion_plus_30_40,
            "40_50": pion_plus_40_50, "50_60": pion_plus_50_60, "60_70": pion_plus_60_70,
            "70_80": pion_plus_70_80, "80_92": pion_plus_80_92
        },
        321: {  # K⁺
            "0_5": kaon_plus_0_5, "5_10": kaon_plus_5_10, "10_15": kaon_plus_10_15,
            "15_20": kaon_plus_15_20, "20_30": kaon_plus_20_30, "30_40": kaon_plus_30_40,
            "40_50": kaon_plus_40_50, "50_60": kaon_plus_50_60, "60_70": kaon_plus_60_70,
            "70_80": kaon_plus_70_80, "80_92": kaon_plus_80_92
        },
        2212: {  # p
            "0_5": proton_plus_0_5, "5_10": proton_plus_5_10, "10_15": proton_plus_10_15,
            "15_20": proton_plus_15_20, "20_30": proton_plus_20_30, "30_40": proton_plus_30_40,
            "40_50": proton_plus_40_50, "50_60": proton_plus_50_60, "60_70": proton_plus_60_70,
            "70_80": proton_plus_70_80, "80_92": proton_plus_80_92
        }
    }

    # 粒子信息
    particle_info = {
        211: {'symbol': r'$\pi^+$', 'name': 'pion'},
        321: {'symbol': r'$K^+$', 'name': 'kaon'},
        2212: {'symbol': r'$p$', 'name': 'proton'}
    }
    
    # 默认缩放因子 (如果没有提供)
    if scale_factors is None:
        scale_factors = [1.0 / (i+1) for i in range(len(centralities))]
    elif len(scale_factors) != len(centralities):
        raise ValueError("scale_factors长度必须与centralities相同")

    # 创建图形
    plt.figure(figsize=(8, 6))
    
    # 获取粒子信息
    particle = particle_info.get(pid, {'symbol': f'PID{pid}', 'name': f'PID{pid}'})
    
    # 标记样式循环
    markers = ['o', 's', '^', 'D', 'v', '<', '>', 'p', '*', 'h', 'X']
    colors = ['red', 'blue', 'green', 'orange', 'black', 'purple', 'cyan', 'magenta', 'lime', 'brown','pink']

    
    # 遍历每个中心度
    for idx, cent in enumerate(centralities):
        scale = scale_factors[idx]
        m, n = map(int, cent.split('_'))
        
        # 获取实验数据
        try:
            exp_data = exp_data_mapping[pid][cent]
            
            # 绘制实验数据 (带误差棒)
            plt.errorbar(
                x=exp_data['pT'],
                y=exp_data['yield'] * scale,
                yerr=[-exp_data['stat_err_minus'] * scale, exp_data['stat_err_plus'] * scale],
                fmt=markers[idx],
                color=colors[idx],
                markersize=6,
                capsize=3,
                alpha=0.8,
                label=f'PHENIX {m}-{n}%' if idx == 0 else ""
            )
        except KeyError:
            print(f"Warning: No experimental data for PID {pid} at centrality {cent}")
            continue
        
        # 获取模拟数据
        sim_data = get_pt_spectrum_data(
            m=m, n=n,
            pid=pid,
            nsampls=2000,
            pt_range=pt_range
        )
        
        # 绘制模拟数据
        plt.plot(
            sim_data['bin_centers'],
            sim_data['bin_means'] * scale,
            color=colors[idx],
            linestyle='-',
            linewidth=2,
            alpha=0.8,
            label=f'CLVisc {m}-{n}%' if idx == 0 else ""
        )
        
#         # 在曲线上方添加标注
#         mid_idx = len(sim_data['bin_centers']) // 2
#         plt.text(
#             sim_data['bin_centers'][mid_idx],
#             sim_data['bin_means'][mid_idx] * scale * 1.2,
#             f"{m}-{n}% ×{scale}",
#             color=colors[idx],
#             fontsize=10,
#             ha='center',
#             va='bottom',
#             rotation=-5,       # 旋转5度（可改为任意角度，如30、45等）
#         )

        # 在曲线上方添加标注 - 修改为科学计数法显示
        mid_idx = len(sim_data['bin_centers']) // 2
        exponent = int(np.log10(scale)) if scale != 0 else 0
        scale_text = f"$10^{{{exponent}}}$" if exponent != 0 else "1"
        
        plt.text(
            sim_data['bin_centers'][mid_idx],
            sim_data['bin_means'][mid_idx] * scale * 1.2,
            f"{m}-{n}% ×{scale_text}",
            color=colors[idx],
            fontsize=10,
            ha='center',
            va='bottom',
            rotation=-10,
        )

    # 创建自定义图例
    legend_elements = [
        Line2D([0], [0], color='black', lw=2, label='CLVisc'),
        Line2D([0], [0], marker='o', color='black', label='PHENIX',
               markersize=8, linestyle='None')
    ]
    plt.legend(handles=legend_elements, fontsize=16, framealpha=0.9, loc='best')

    # 设置图形属性
    plt.xlim(pt_range)
    if ylog:
        plt.yscale('log')
    
    plt.xlabel(r'$p_T$ [GeV]', fontsize=16,ha='center')
    plt.ylabel(r'$d^2N/2\pi p_T dp_T dy$ [GeV$^{-2}$]', fontsize=16,ha='center')
    plt.title(rf"{particle['symbol']} in Au+Au $\sqrt{{s_{{\rm NN}}}}$=200 GeV", fontsize=20,ha='center')
    
#     plt.grid(True, which='both', linestyle=':', alpha=0.3)
    plt.tight_layout()
    plt.show()
```
